### Important!!!

I noticed there's an outliers in the scatter plot above, i removed the outliers from jamsCount and jamsDelay and it visualized better now, below

```python
df2 = df1
```

```python
def remove_outliers(df, column):
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    IQR = q3 - q1 
    lower_bound = q1 - 1.5 * IQR
    upper_bound = q3 + 1.5 * IQR

    return df[(df[column]>= lower_bound) & (df[column]<= upper_bound)]
df2 = remove_outliers(df2, 'JamsCount')
df2 = remove_outliers(df2, 'JamsDelay')
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(data=df2, x='JamsCount', y ='JamsDelay')
plt.title('Relationships between JamsCount and JamsDelay')
plt.xlabel('JamsCount')
plt.ylabel('JamsDelay')
plt.show()
```
